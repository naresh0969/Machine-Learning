```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Week7_train.csv')
df
```

```python
df_sample=df.sample(4200)
df_sample
```

```python
y=df_sample['label']
X=df_sample.drop(columns=['label'],axis=1)
```

```python
print(y.shape)
print(X.shape)

```

```python
import matplotlib.pyplot as plt
plt.imshow(X.iloc[0].values.reshape(28,28))
```

```python
#converting into 1d array
y=y.values.ravel()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
```

```python
test_sample=pd.read_csv('Week7_test.csv')
test_sample=test_sample.sample(2800)
test_sample.shape
```

```python
X
```

```python
y_train_pred=model.predict(X)
print(classification_report(X,y_train_pred))
```
