```python
#6.  Prepare X (feature variables) and y (response variable)

import pandas as pd

# Load the dataset
df = pd.read_csv('tvmarketing.csv')

# Prepare the feature variable (X) and the response variable (y)
X = df['TV']  # Feature variable
y = df['Sales']  # Response variable

# Print the shapes to verify
print('Shape of X:', X.shape)
print('Shape of y:', y.shape)


```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the resulting datasets to verify
print('Shape of X_train:', X_train.shape)
print('Shape of X_test:', X_test.shape)
print('Shape of y_train:', y_train.shape)
print('Shape of y_test:', y_test.shape)
```

```python
#9
import numpy as np

# Calculate means of X_train and y_train
X_mean = X_train.mean()
y_mean = y_train.mean()

# Calculate the numerator and denominator for the slope (β1)
numerator = np.sum((X_train - X_mean) * (y_train - y_mean))
denominator = np.sum((X_train - X_mean) ** 2)

# Calculate slope (β1) and intercept (β0)
beta_1 = numerator / denominator
beta_0 = y_mean - (beta_1 * X_mean)

# Print the coefficients
print(f"Intercept (β0): {beta_0}")
print(f"Slope (β1): {beta_1}")
```

```python
#using library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('tvmarketing.csv')

# Prepare the feature variable (X) and the response variable (y)
X = df['TV']


print(f"X shape {X.shape}")
y = df['Sales']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the Linear Regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train.values.reshape(-1, 1), y_train)

# Print the coefficients
print(f"Intercept : {model.intercept_}")
print(f"Slope : {model.coef_[0]}")

# Make predictions on the test set
y_pred = model.predict(X_test.values.reshape(-1, 1))

# Print the predicted and actual values
print("Predicted values for y (Sales):")
print(y_pred[:5])  # First 5 predictions

print("\nActual values for y (Sales):")
print(y_test[:5])  # First 5 actual values

custom_tv_value=156
predicted_sales = model.predict([[custom_tv_value]])
print(f"predicted_sales{predicted_sales}")

```

```python
#16. Print Mean Squared Error and R Squared Value
import numpy as np

# Function to calculate Mean Squared Error (MSE)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Function to calculate R-squared (R²)
def r_squared(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)

# Calculate predictions on the test set
y_pred = predict(X_test, beta_0, beta_1)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate R-squared (R²)
r2 = r_squared(y_test, y_pred)
print(f"R-squared (R²): {r2}")
```

```python
#17. Calculate and print R Squared and Adjusted R Squared values without any library 
import numpy as np

# Function to calculate R-squared (R²)
def r_squared(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)  # Total sum of squares
    ss_residual = np.sum((y_true - y_pred) ** 2)        # Residual sum of squares
    return 1 - (ss_residual / ss_total)

# Function to calculate Adjusted R-squared (R² adj)
def adjusted_r_squared(y_true, y_pred, X):
    r2 = r_squared(y_true, y_pred)
    n = len(y_true)   # Number of data points
    k = X.shape[1]    # Number of independent variables
    return 1 - ((1 - r2) * (n - 1)) / (n - k - 1)

# Assuming the function to predict is defined as follows
def predict(X, beta_0, beta_1):
    return beta_0 + beta_1 * X

# Example coefficients after linear regression (replace with actual values)
beta_0 = 2  # Example intercept
beta_1 = 0.05  # Example slope

# Calculate predictions on the test set
y_pred = predict(X_test, beta_0, beta_1)

# Calculate R-squared (R²)
r2 = r_squared(y_test, y_pred)
print(f"R-squared (R²): {r2}")

# Calculate Adjusted R-squared (R² adj)
r2_adj = adjusted_r_squared(y_test, y_pred, X_test.reshape(-1, 1))
print(f"Adjusted R-squared (R² adj): {r2_adj}")
```
