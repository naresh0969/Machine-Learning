```python
#6.  Prepare X (feature variables) and y (response variable)
X=df['TV']
y=df['Sales']
```

```python
type(X)
```

```python
#7.  Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test 

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)


```

```python
X_train
```

```python
#9.  Perform Linear Regression (on train data) without using any library and print the coefficients values
X_mean=X_train.mean()
y_mean=y_train.mean()

numerator=np.sum((X_mean-X_train)*(y_train-y_mean))
denominator=np.sum((X_mean-X_train)**2)

m=numerator/denominator
c=y_mean-(m*X_mean)
print("Intercept ",m)
print("Slop ",c)
```

```python
#10. Create the model using the calculated coefficients and make predictions on the testing set

a_pred=m*X_test+c
a_pred
```

```python
X_test.shape
a_pred
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,a_pred)
```

```python
X_train=np.array(X_train).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)
X_test=np.array(X_test).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)

```

```python

#11. Perform Linear Regression using library(sklearn) 
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(X_train,y_train)
```

```python
#12. Print the intercept and coefficients 
print(model.coef_)
print(model.intercept_)
print(model.score(X_train,y_train))
```

```python
#13. Make predictions on the testing set - y_pred 
X_pred=model.predict(X_test)
y_pred=model.predict(y_test)
```

```python
print(X_test.shape)
print(y_test.shape)
print(y_pred.shape)
```

```python
#14. Plot train data and the model
plt.scatter(X_test,y_test)
plt.plot(X_test,X_pred)
```

```python
#16. Print Mean Squared Error and R Squared Value
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)  # small value → good
r2 = r2_score(y_test, y_pred)             # close to 1 → good
print("MSE:", mse)
print("R2:", r2)
```
