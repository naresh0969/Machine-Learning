```python
#11.Sort the data w.r.t price_find the details of the most and the least expensive cars
import pandas as pd
df=pd.read_csv("/home/user/Downloads/Car_sales.csv")

# Sort the data by price (ascending)
df_sorted = df.sort_values(by='Price in thousands', ascending=True)

# Find the least expensive car
least_expensive_car = df_sorted.iloc[0]

# Find the most expensive car
most_expensive_car = df_sorted.iloc[-1]

# Print the results
print("Details of the Least Expensive Car:")
print(least_expensive_car)

print("\nDetails of the Most Expensive Car:")
print(most_expensive_car)
```

```python
#14. Plot histogram of continuous numerical variable : price,sales,hoursepower,fuelefficiency
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv("/home/user/Downloads/Car_sales.csv")  # Replace with actual file path
df.replace([np.inf, -np.inf], np.nan, inplace=True)
# List of continuous numerical columns
columns = ["Price in thousands", "Sales in thousands", "Horsepower", "Fuel efficiency"]

# Create histograms
plt.figure(figsize=(12, 8))

for i, col in enumerate(columns, 1):
    plt.subplot(2, 2, i)  # 2 rows, 2 columns
    sns.histplot(df[col], bins=30, kde=True)  # KDE=True adds a density curve
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
