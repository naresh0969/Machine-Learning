```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("newsgroups_train.csv")
df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorize = CountVectorizer(max_features=1500)
vectors = vectorize.fit_transform(df['text'])
vectors = vectors.toarray()
vectors
```

```python
for vec in vectors[0]:
    print(vec, end=" ")
```

```python
vectors.shape
```

```python
X = vectors
y = df.target
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```

```python
model = MultinomialNB()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

predictions
```

```python
# Assuming y_test and predictions are iterables of the same length
for y, pred in zip(y_test, predictions):
    print(f"y_test: {y}, predictions: {pred}")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
cm = confusion_matrix(y_test,predictions)
cm
```

```python
cr = classification_report(y_test,predictions)
print(f"cr: {cr}")
accuracy = accuracy_score(y_test,predictions)
print(f"accuracy: {accuracy}")
```
